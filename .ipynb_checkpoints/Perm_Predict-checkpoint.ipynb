### Porosity Permeability Relationship

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
core_data = pd.read_csv("Data/Well_Core.csv", na_values=' ')
```

```python
core_data.head()
```

```python
core_data.describe()
```

Scatter plot:

```python
core_data.plot(kind="scatter", x="CPOR", y="CPERM")
```

```python
core_data['CPOR'].plot(kind='box');
```

There is one main problem with previous plot. Poro-perm data is normally presented on semilog scale, where the y-axis is logarithmic and the x-axis set to linear. We can change this by using the yscale('log') method.

```python
core_data.plot(kind="scatter", x="CPOR", y="CPERM")
plt.yscale('log')
plt.grid(True)
```

We can further refine the scatter plot by removing the scientific notation on the y-axis.

```python
from matplotlib.ticker import FuncFormatter
fig, ax = plt.subplots()

ax.axis([0, 0.25, 0.03, 1000])
ax.plot(core_data['CPOR'], core_data['CPERM'], 'bo')
ax.set_yscale('log')
ax.grid(True)
ax.set_ylabel('Core Perm (mD)')
ax.set_xlabel('Core Porosity')

#Format the axes so that they show whole numbers
for axis in [ax.yaxis, ax.xaxis]:
    formatter = FuncFormatter(lambda y, _: '{:.16g}'.format(y))
    axis.set_major_formatter(formatter)
```

We can derive a linear regression by using the polyfit function from numpy. As we are working with a semi-log plot, we need to calculate log base 10 of the core permeability.

```python
clean_data = core_data.dropna()
```

```python
core_data.dropna()
```

```python
clean_data.head()
```

```python
import statsmodels.api as sm

x = clean_data['CPOR']
y = np.log10(clean_data['CPERM'])

model = sm.OLS(y, x, missing='drop')
results = model.fit()
```

```python
results.summary()
```

```python
x = np.polyfit(clean_data['CPOR'], np.log10(clean_data['CPERM']), 1)
```

```python
x
```

The result of the regression is: $10^{(23.35642315  * CPOR -2.78915149)}$

Using the code from the previous plot, we can now add a new line with our x-axis set to porosity and our y-axis set to an equation. Note that we have to use the equation above to reverse the log base 10 function used when deriving the function.

```python
from matplotlib.ticker import FuncFormatter
fig, ax = plt.subplots()

ax.axis([0, 0.25, 0.01, 1000])
ax.semilogy(clean_data['CPOR'], clean_data['CPERM'], 'bo')

ax.grid(True)
ax.set_ylabel('Core Perm (mD)')
ax.set_xlabel('Core Porosity')

ax.semilogy(clean_data['CPOR'], 10**(x[0] * clean_data['CPOR'] + x[1]), 'r-')


#Format the axes so that they show whole numbers
for axis in [ax.yaxis, ax.xaxis]:
    formatter = FuncFormatter(lambda y, _: '{:.16g}'.format(y))
    axis.set_major_formatter(formatter)
    
```

Our next step is to calculate a predicted permeability using our new equation and add it as a new column in the pandas dataframe.

```python
clean_data.loc[:, "PRED_PERM"] = 10**(x[0] * clean_data['CPOR'] + x[1])
```

```python
clean_data.describe()
```

We can then make a quick plot of our prediction vs the original measurement. To get an idea of how well the prediction did, we can add a 1 to 1 line to aid the visualisation.

```python
fig, ax = plt.subplots()

ax.axis([0.01, 1000, 0.01, 100])
ax.loglog(clean_data['CPERM'], clean_data['PRED_PERM'], 'bo')
ax.loglog([0.01, 1000], [0.01, 100], 'r-')
```

```python
clean_data.plot(kind="scatter", x="PHIZ", y="RQI")
plt.yscale('log')
plt.grid(True)
```

```python
from sklearn.model_selection import train_test_split
X = clean_data[['PHIZ', 'RQI']]
y = clean_data['CPERM']
```

```python
from sklearn import linear_model
reg = linear_model.LinearRegression()
```

```python
reg.fit (X, y)
# The coefficients
print ('Coefficients: ', reg.coef_)
print ('Intercept: ', reg.intercept_)
```
